```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2, 100)

plt.plot(x, x, label = 'LINEAR')
plt.plot(x, x**2, label = 'QUADRATICO')
plt.plot(x, x**3, label = 'CUBICO')

plt.xlabel('Horizontal')
plt.ylabel('Vertical')
plt.title('Grafico de crescimento')
plt.show()
```
